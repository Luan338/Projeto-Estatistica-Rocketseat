```python
## Trazer a média das vendas
## Criar um gráfico de barras vertical mostrando o mês de referência e o valor
## Criar um gráfico de linhas mostrando o mês de referência e o valor.r.
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Dicionário de faturamento
dict_faturamento = {
    'data_ref': [
        '2023-01-01', 
        '2020-02-01', 
        '2021-03-01', 
        '2022-04-01', 
        '2023-05-01',
        '2023-06-01', 
        '2020-07-01', 
        '2021-08-01', 
        '2022-09-01', 
        '2023-10-01',
        '2022-11-01', 
        '2023-12-01',
        ],
    'valor': [
        400000, 
        890000, 
        760000, 
        430000, 
        920000,
        340000, 
        800000, 
        500000, 
        200000, 
        900000,
        570000, 
        995000,
        ]
}
```

```python
## Transformando em um DataFrame
df_faturamento = pd.DataFrame.from_dict(dict_faturamento)
```

```python
## MÉDIA DAS VENDAS

f"A Média é de: {df_faturamento['valor'].mean():.2f}"
```

```python
# Gráfico de barras vertical mostrando o mês de referência e o valor

## Colocando os valores em uma variável
Data = df_faturamento['data_ref']
Valor = df_faturamento['valor']

# Convertendo 'data_ref' para datetime
df_faturamento['data_ref'] = pd.to_datetime(df_faturamento['data_ref'], format='%Y-%m-%d') 

# Ordenando pela coluna 'data_ref' em ordem crescente
df_faturamento = df_faturamento.sort_values(by='data_ref', ascending=False)

Data = df_faturamento['data_ref'].dt.strftime('%d/%m/%Y')
Valor = df_faturamento['valor']

## Plotando
fig, ax = plt.subplots()

## Criando o gráfico de barras vertical (troque os eixos)
bars = ax.bar(Data, Valor, width=0.6, edgecolor="white", linewidth=0.7)

## Rótulos dos eixos
ax.set_title('Comparação dos valores por mês')
ax.set_xlabel('Mês de Referência', fontsize=10)
ax.set_ylabel('Valor (R$)', fontsize=10)

## Ajusta o layout para melhor visualização dos rótulos
plt.xticks(rotation=90, ha='right', fontsize=8)
plt.tight_layout()

## Inserir os valores em cada barra
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 50, f'R$ {yval:.2f}', ha='center', va='bottom', fontsize=8)

plt.show()
```

```python
# Criar um gráfico de linhas mostrando o mês de referência e o valor.

# Plot
fig, ax = plt.subplots(figsize=(8, 5))

# Crie o gráfico de linhas
ax.plot(Data, Valor, marker='o', linestyle='-', color='b')

# Rótulos dos eixos
ax.set_title('Progresso de valores por mês')
ax.set_xlabel('Mês de Referência')
ax.set_ylabel('Valor (R$)')

# Ajusta o layout para melhor visualização dos rótulos
plt.xticks(rotation=90, ha='right', fontsize=8)
plt.tight_layout()

# Adicione os valores do eixo Y em cada ponto, acima do marcador
for i, v in enumerate(Valor):
    ax.text(Data[i], v, f'R$ {v:.2f}', ha='center', va='bottom', fontsize=8)

plt.show()
```
